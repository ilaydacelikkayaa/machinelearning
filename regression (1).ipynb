#Kümeleme


```python
 import pandas as pd
df=pd.read_csv("Live.csv")
df.head()
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(df)
```

```python
labels=model.predict(df)
import numpy as np
np.unique(labels,return_counts=True)
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
silhouettes=[]
ks=list(range(2,12))
for n_cluster in ks:
  kmeans=KMeans(n_clusters=n_cluster).fit(df)
  label=kmeans.labels_
  sil_coeff=silhouette_score(df,label,metric='euclidean')
  print('for n_clusters={}, the silhouette coeffivient is {}'.format(n_cluster,sil_coeff))
  silhouettes.append(sil_coeff)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.scatter(ks,silhouettes,marker='x',c='r')
plt.plot(ks,silhouettes)
plt.xlabel('k')
plt.ylabel('silhouette score')

```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4)
model.fit(df)
labels=model.predict(df)
import numpy as np
np.unique(labels,return_counts=True)
```

```python
df['labels']=labels
df['labels'].value_counts()
```

#Regularization

```python
import pandas as pd
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
data=pd.read_csv("train (1).csv")
X=data.drop('SalePrice',axis=1)
y=data.loc[:,'SalePrice']
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
linear_reg=LinearRegression()
ridge_reg= Ridge(alpha=0.05)
#Ridge regresyonu, doğrusal regresyonun bir çeşididir ve
#fazla uydurmayı (overfitting)
# önlemek için bir regularization (düzenleme) parametresi ekler.
```

```python
linear_reg.fit(X_train,y_train)
ridge_reg.fit(X_train,y_train)
linear_pred=linear_reg.predict(X_test)
ridge_pred=ridge_reg.predict(X_test)
```

```python
linear_mse=mean_squared_error(y_test,linear_pred)
ridge_mse=mean_squared_error(y_test,ridge_pred)
print(f'MSE without Ridge:{linear_mse}')
print(f'MSE with Ridge:{ridge_mse}')
#Hata Ölçüsü: MSE, bir modelin tahmin doğruluğunu ölçmek için kullanılır.
# Küçük bir MSE değeri, modelin daha doğru tahminler yaptığını gösterir.
```

#NAIVE BAYES

CONDITIONAL PROBABILITY                                  
P(B|A)= P(A and B)/P(A)                                        
BAYES' THEOREM                                  
P(A|B)= (P(B|A)P(A))/P(B)

```python
import pandas as pd
```

```python
df=pd.read_csv("/breast-cancer (2).csv")
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
X=df.drop('diagnosis',axis=1)
y=df.loc[:,'diagnosis']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)  # cv=5: 5 katmanlı çapraz doğrulama
print("Cross-Validation Scores:", scores)
print("Mean CV Accuracy:", scores.mean())
```

#Decision trees

```python
import pandas as pd
df=pd.read_csv("/content/DecisionTrees_titanic.csv")
df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X=df.drop('Survived',axis=1)
y=df.loc[:,'Survived']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
decisiontree=DecisionTreeClassifier()
randomforest=RandomForestClassifier()

```

```python
decisiontree=DecisionTreeClassifier()
randomforest=RandomForestClassifier()
```

```python
decisiontree.fit(X_train,y_train)
randomforest.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,10))
plot_tree(decisiontree,
          max_depth=2,
          feature_names=X.columns,
          filled=True,
          impurity=True,
          rounded=True,
          precision=1)
#max_depth=2, ağacın sadece 2 katman derinliğinde olacağını belirtir.
#Yani, ağacın sadece kök düğümü ve iki seviyeli alt düğümleri olacak. 3 katman
#feature_names=X.columns herbir düğümün kullanılan özellikleri
# Impurity, karar ağacının her bir düğümünde veri karışıklığını ölçer.
#precision=1, sayıları bir basamağa kadar gösterir.
# Örneğin, impurity değeri 0.5 veya 0.42 olarak gösterilebilir.
plt.show()
```

```python
decisiontree_pred=decisiontree.predict(X_test)
randomforest_pred=randomforest.predict(X_test)
```

```python
print(classification_report(y_test,decisiontree_pred))
print(classification_report(y_test,randomforest_pred))
```
